```python
#This takes in the date/time/hour/zip data and creates seperates feasutres 

import pandas as pd
import numpy as np
from datetime import datetime, date, time
```

```python
df = pd.read_csv('demand_data_2017.csv', index_col='time_block')

```

```python
df.head
```

```python
# First, remove the first row as its weird
df = df.dropna()
```

```python
df
```

```python
#First, Isolate the Hour
blocks = df.index
Hours = [0] * len(df)
Days = [0] * len(df)
Days_o_week = [0] * len(df)
Months = [0] * len(df)

for i in range(len(df)):
    #Get the Hours
    current_block = blocks[i]
    Hour = current_block[-2:]
    Hours[i] = int(Hour)
    
    #Get the Dates
    date = current_block[:10]
    date = datetime.strptime(date, '%Y/%m/%d')
    
    Days[i] = date.day
    Days_o_week[i] = date.weekday()
    Months[i] = date.month
    
    
    

#Merge the Times into the Data frame
df['Hour'] = Hours
df['Day'] = Days
df['Day_of_week'] = Days_o_week
df['Month'] = Months
print(df)
print(type(df))
```

```python
#Export to csv
df.to_csv('Demand_Data2_2017.csv', sep=',')
```
