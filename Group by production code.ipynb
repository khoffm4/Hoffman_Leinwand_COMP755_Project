```python
import pandas as pd
data = pd.read_csv('yellow_cab.csv.gz',  compression='gzip',
                   error_bad_lines=False)


```

```python
data.head
```

```python
#grouping by time_block, pickup_zip and dropoff_zip. Everything kept for demand data will be summed (later turned to averages if used for neighbors data). Anything used only for neighbors data can be averaged immediately.  
#passenger_count count represents the number of trips at that time block from pickup_zip to dropoffzip. The sum tells the total number of passengers
first_grouping=data.groupby(['time_block', 'pickup_zip', 'dropoff_zip'],  as_index=False).agg({'passenger_count': ['count', 'sum'], 'trip_distance': 'mean', 'tip_amount': 'sum', 'tolls_amount': 'sum', 'total_amount': 'sum',   'time_delta':'mean' })

first_grouping.to_csv('first_grouping', encoding='utf-8')
```

```python
#grouping even further, ignoring pickup location and only worrying about demand
demand_data=first_grouping.groupby(['time_block', 'pickup_zip'],  as_index=False).agg({list(first_grouping)[3]: 'sum', list(first_grouping)[4]: 'sum', list(first_grouping)[6]: 'sum',  list(first_grouping)[7]: 'sum', list(first_grouping)[8]: 'sum' })

demand_data.to_csv('demand_data', encoding='utf-8')
```
