```python
######################################
######################################
######################################
######################################
#CREATING NEW DEMAND DATASET WITH ALL ZEROS INCLUDED
#######################################
######################################
#######################################
######################################
import pandas as pd
demand_2017= pd.read_csv('demand_data_2017.csv')
```

```python
demand_2017.shape
demand_2017= demand_2017.iloc[1:, 1:4]
demand_2017.head()
```

```python
demand_2017[['time_block', 'PULocationID']]
```

```python
zerodata = demand_2017.groupby(['time_block','PULocationID']).sum().unstack(fill_value=0).stack()
```

```python
zerodata.head
```

```python
___, month, day_of_month, hour = zip(*zerodata['time_block'].apply(lambda x: x.replace(' ','/').split('/')))
```

```python
zerodata.to_csv('demand_data_2017_with_zeros.csv', encoding='utf-8')
```

```python
######################################
######################################
######################################
######################################
#CREATING NEW NEIGHBORS DATASET WITH ALL ZEROS INCLUDED
#######################################
######################################
#######################################
######################################
neighbors_2017= pd.read_csv('neighbors_data_2017_fixed.csv')
```

```python
neighbors_2017.head
```

```python
neighbors_2017.head()
```

```python
#keepcolumns=(1,2,3, 11,12, 13,14,15,16)
neighbors_2017=neighbors_2017.iloc[1:,]
neighbors_2017=neighbors_2017[['time_block', 'PU_DO', 'time_delta',  'month',  'hour',  'day_of_month',  'day_of_week'] ]
```

```python
neighbors_2017.head
```

```python
neighbors_2017['time_delta']=pd.to_numeric(neighbors_2017['time_delta'])
```

```python
zerodata2 = neighbors_2017.groupby(['time_block', 'PU_DO']).sum().unstack(fill_value=0).stack()
```

```python
zerodata2.head
```

```python
zerodata2.to_csv('neighbors_data_2017_fixed_with_zeros.csv', encoding='utf-8')
```

```python
zerodata3= pd.read_csv('neighbors_data_2017_fixed_with_zeros.csv')
```

```python
zerodata3.head
```

```python
___, month, day_of_month, hour = zip(*zerodata3['time_block'].apply(lambda x: x.replace(' ','/').split('/')))
```

```python
month = list(month)
hour = list(hour)
day_of_month = list(day_of_month)

zerodata3['month'] = month
zerodata3['hour'] = hour
zerodata3['day_of_month'] = day_of_month
```

```python
from datetime import datetime as dt

dow = zerodata3['time_block'].apply(lambda x: dt.strptime(x, '%Y/%m/%d %H').weekday())
#display(adj_grouping_for_2017.head()['time_block'].apply(lambda x: dt.strptime(x, '%Y/%m/%d %H')))
zerodata3['day_of_week'] = dow
```

```python
zerodata3.head
```

```python
zerodata3.to_csv('neighbors_data_2017_fixed_with_zeros.csv', encoding='utf-8')
```
