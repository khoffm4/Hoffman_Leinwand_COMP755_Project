```python
datetime_list = [0] * pickuptimes.shape[0]
for i in range(pickuptimes.shape[0]):
    datetime_list[i] =  datetime.strptime(pickuptimes[i][0], '%Y-%m-%d %H:%M:%S')

```

```python
print(datetime_list[1].year)
```

```python
time_mat = np.arange(pickuptimes.shape[0] * 4).reshape(pickuptimes.shape[0], 4)
print(time_mat)
```

```python

for i in range(pickuptimes.shape[0]):
    time_mat[i,0] = datetime_list[i].month
    time_mat[i,1] = datetime_list[i].day
    time_mat[i,2] = datetime_list[i].hour
    time_mat[i,2] = str(datetime_list[0].year) + '/' + str(datetime_list[0].month) + '/' + str(datetime_list[0].day) + '/' + str(datetime_list[0].hour)
    
    
```

```python
print(time_mat)
```

```python
np.savetxt("time_mat.csv", time_mat, delimiter=",")
```
